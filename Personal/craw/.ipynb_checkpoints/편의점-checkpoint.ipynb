## 세븐일레븐

```python
with open('./cvs/data/seveneleven_all.json', 'r', encoding='utf-8') as f:
    json_data = json.loads(f.read())
```

```python
sevenEleven = json_data['store_list']
```

```python
sevenEleven[0]
```

```python
s_name = []
address = []
lat = []
lot = []
```

```python
for i in range(len(sevenEleven)):
    s_name.append(sevenEleven[i]['s_name'])
    address.append(sevenEleven[i]['doro_address'])
    lat.append(sevenEleven[i]['lat'])
    lot.append(sevenEleven[i]['lot'])
```

```python
cvs_se = pd.DataFrame({'s_name':s_name, 'address':address, 'lat':lat, 'lot':lot})
```

```python
cvs_se['s_brand'] = '세븐일레븐'
```

```python
cvs_se
```

## CU

```python
with open('./cvs/data/cu01.json', 'r', encoding='utf-8') as f:
    json_data1 = json.loads(f.read())
with open('./cvs/data/cu02.json', 'r', encoding='utf-8') as f:
    json_data2 = json.loads(f.read())
with open('./cvs/data/cu03.json', 'r', encoding='utf-8') as f:
    json_data3 = json.loads(f.read())
with open('./cvs/data/cu04.json', 'r', encoding='utf-8') as f:
    json_data4 = json.loads(f.read())
```

```python
cu1 = json_data1['store_list']
cu2 = json_data2['store_list']
cu3 = json_data3['store_list']
cu4 = json_data4['store_list']
```

```python
cu1[0]
```

```python
s_name = []
address = []
tel = []
```

```python
for i in range(len(cu1)):
    s_name.append(cu1[i]['s_name'])
    address.append(cu1[i]['doro_address'])
    tel.append(cu1[i]['tel'])
for i in range(len(cu2)):
    s_name.append(cu2[i]['s_name'])
    address.append(cu2[i]['doro_address'])
    tel.append(cu2[i]['tel'])
for i in range(len(cu3)):
    s_name.append(cu3[i]['s_name'])
    address.append(cu3[i]['doro_address'])
    tel.append(cu3[i]['tel'])
for i in range(len(cu4)):
    s_name.append(cu4[i]['s_name'])
    address.append(cu4[i]['doro_address'])
    tel.append(cu4[i]['tel'])
```

```python
cvs_cu = pd.DataFrame({'s_name':s_name, 'address':address, 'tel':tel})
```

```python
cvs_cu['s_brand'] = '씨유'
```

```python
cvs_cu.head(20)
```

```python
cvs = pd.merge(cvs_cu, cvs_se, how='outer')
```

```python
cvs
```
